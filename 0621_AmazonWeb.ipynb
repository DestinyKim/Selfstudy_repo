### 전체적인 리뷰 확인하기

```python
all_page = soup.find_all("div", class_="a-section celwidget")[1]

one_pro = all_page.find("a", class_="a-size-mini a-link-normal a-color-secondary").text
one_pro
```

```python
import time

all_user = [] # 사용자
all_dates = [] # 날짜
all_product = [] # 상품형태
all_ratings = [] # 평점
all_reviews = [] # 리뷰

for page in range(2, 7, 1):
    time.sleep(3)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    all_page = soup.find_all("div", class_="a-section celwidget")

    for one in all_page:
        # 사용자
        review_one = one.find("span", class_="a-profile-name").text
        all_user.append(review_one)

        # 날짜
        date_one = one.find("span", class_="a-size-base a-color-secondary review-date")
        texts = date_one.text.split("on")
        date = texts[1].strip()
        all_dates.append(date)
        
        # 상품형태
        one_pro = one.find("a", class_="a-size-mini a-link-normal a-color-secondary").text
        all_product.append(one_pro)

        # 평점
        raiting_one =  one.find("span", class_="a-icon-alt").text
        nums = re.findall('\d+', raiting_one)[0]
        all_ratings.append(nums)

        # 리뷰
        review_one = one.find("span", class_="a-size-base review-text review-text-content")
        tmp = review_one.text
        review = tmp.strip()
        all_reviews.append(review)
    
    # 확인
    print("user : " , all_user[-1],"/","rating :", all_ratings[-1])
    print("shape :", all_product[-1])
    print("review : ", all_review[-1], end="\n\n\n")
    
    # 다음 페이지 클릭
    sel_next = driver.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
    sel_next.click()   
```

```python
import pandas as pd
from pandas import DataFrame

# df = df(data = {"user":all_user, "date":all_dates, "rate": all_ratings, "review":all_reviews}, 
#         columns=['User','Date','Rate','Review'])
# df

data = {"user":all_user, "date":all_dates,"shape":all_product, "rate": all_ratings, "review":all_reviews}
dat = pd.DataFrame(data)
dat
```
