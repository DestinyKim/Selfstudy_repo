```python
def get_day_review(movie_code, page_no):
    
    url = f"https://movie.naver.com/movie/board/review/list.naver?st=code&sword={movie_code}&target=after&page={page_no}"
    headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36"}    
    response = requests.get(url, headers=headers)
    html = bs(response.text, "lxml")
    tables = html.select("table")
    table = pd.read_html(str(tables))
    df_page = table[1].dropna(how="all", axis=1)
    
    return df_page
```

```python
get_day_review("191597", 1)
```
