```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
movie_url= 'https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=189150&target=after'
page = urlopen(movie_url)
soup = BeautifulSoup(page, "html.parser")
comment_all = soup.find_all('td', class_='title')
list(comment_all)
```

```python
print(comment_all[1].prettify())
```

```python
com_one = list(comment_all[1].children)[6].strip()
```

```python
comments =[]
for i in range(1,51):
    url = movie_url + str(i)
    page = urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    
    comment_all = soup.find_all('td', class_='title')
    
    for one in comment_all:
        one_com = list(one.children)[6].strip()
        print(one_com)
        comments.append(one_com)
```

```python
import pandas as pd
import os
```

```python
dict_doc = {"text":comments}
doc = pd.DataFrame(dict_doc)
doc.to_csv("분노의질주_더얼티메이트리뷰_0617.csv", index=False)

os.listdir(os.getcwd())
```
