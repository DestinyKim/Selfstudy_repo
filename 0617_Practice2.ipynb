# 네이버 영화 정보 가져오기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://movie.naver.com/movie/running/current.nhn"
page = urlopen(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
soup.find("div",class_="obj_section").find_next("dt",class_="tit").find_next("a").text
```

```python
# tit = list(soup.find_all("dt",class_="tit"))
tit=list(soup.find("dt",class_="tit").children)[3]
tit
```

## 01 제목가져오기

```python
list1 = soup.find_all("dt",class_='tit')
print(list1)
list1[0].find('a').text
```

```python
movie_title = []
for one in list1:
#     print(one.find('a').text)
    one_movie = one.find('a').text
    movie_title.append(one_movie)
movie_title    
```

## 02 네이버 영화 평점 가져오기

```python
grades = soup.find_all("span",class_="num")
grades[0].text
```

```python
movie_grades = []
for idx, grade in enumerate(grades): 
    movie_grade = grade.text
    if idx%2==0:
           movie_grades.append(movie_grade)
    else:
        pass
movie_grades
```

## 03 네이버 영화 참여 인원 가져오기

```python
person = soup.find_all("span",class_='num2')
person[0].text
# people[]
```

```python
people=[]
for one in person:
    person1 = one.text
    people.append(person1)
people
```

## 04 네이버 영화 예매율 가져오기

```python
tickets = soup.find_all("span",class_="num")
ticket_ratio = []
for idx, ticket in enumerate(tickets): 
    movie_ticket = ticket.text
    if idx%2==1:
           ticket_ratio.append(movie_ticket)
    else:
        pass
ticket_ratio
```
